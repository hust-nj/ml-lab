```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

初始化构造数据集

```python
divide = 0.1
dataset = np.random.randn(1000, 2)
dataset = dataset[np.abs(dataset[:,1] - divide) > 1e-1]
label = (dataset[:,1] > divide).astype(int)
label[dataset[:,1] < divide] = -1
above, below = dataset[dataset[:,1] > divide], dataset[dataset[:,1] < divide]
plt.scatter(above[:,0], above[:,1], marker = 'x')
plt.scatter(below[:,0], below[:,1], marker = '.')
```

进行缩放，旋转，平移

```python
angle = 30 * np.pi / 180
mu, sigma = [3, 5], 4
rotate = np.array([[np.cos(angle), np.sin(angle)], [-np.sin(angle), np.cos(angle)]])
above, below, dataset = sigma * above.dot(rotate) + mu, sigma * below.dot(rotate) + mu, sigma * dataset.dot(rotate) + mu

plt.scatter(above[:,0], above[:,1], marker = 'x')
plt.scatter(below[:,0], below[:,1], marker = '.')
```
